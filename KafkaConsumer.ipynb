```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
import time
```

```python

consumer = KafkaConsumer(
    'demo_test',  # Ensure this matches producer topic
    bootstrap_servers=['13.127.163.201:9092'],  # Change IP as needed
    value_deserializer=lambda x: loads(x.decode('utf-8')),
    auto_offset_reset='earliest',  # Start from earliest message
    enable_auto_commit=True
)
```

```python
s3 = S3FileSystem()
```

```python
from json import dump


try:
    for count, message in enumerate(consumer):
        print("Received:", message.value)
        filename = f"s3://kafka-stock-market-analysis-shivam/stock_market_{int(time.time())}_{count}.json"
        with s3.open(filename, 'w') as file:
            dump(message.value, file)
except Exception as e:
    print("Error:", e)
```
